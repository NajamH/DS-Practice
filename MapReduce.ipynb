```python
def map(input_key, input_values):
  result = []
  for word in input_values.split():
    result.append([word,"1"])
  return result
```

```python
def reduce(output_key, intermediate_values):
  result = 0
  for word, key in intermediate_values :
    if word == output_key :
      result = result + 1
  
  return result
```

```python
intermediate_values = map("1", text_values)
print (intermediate_values)
```

```python
intermediate_values.sort()
print (intermediate_values)
```

```python
import numpy as np
getWords = np.array([])
i = 0
for word, x in intermediate_values :
  getWords = np.append(getWords, intermediate_values[i][0])
  i = i+1

distinctWords = np.unique(getWords)

for word in distinctWords :
  print(word, reduce(word, intermediate_values))
  
```
